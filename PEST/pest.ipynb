## json 파일 type '사각형' -> bbox 수정

## yolov5 라벨링 text 파일 만들기

```python
import os
import json
root_dir = 'C:\\Users\\user\\Desktop\\NIA_최종데이터_1207\\해충트랩\\해충\\label\\'
dir_list=os.listdir(root_dir)
d=0
for i in dir_list:
    filepath='C:\\Users\\user\\Desktop\\NIA_최종데이터_1207\\해충트랩\\해충\\images\\'+i.split(".")[0]+'.jpg'
    img = hangulFilePathImageRead(filepath)

    with open('C:\\Users\\user\\Desktop\\NIA_최종데이터_1207\\해충트랩\\해충\\label\\'+i, "r",encoding="utf-8",errors='ignore') as json_file:
        json_data = json.load(json_file)
        x_scale=1/float(json_data["description"]["width"])
        y_scale=1/float(json_data["description"]["height"])
        a_list=[]
#         x_scale=1/float(img.shape[0])
#         y_scale=1/float(img.shape[1])
#         json_data = json.load(json_file)
#         a_list=[]
        
        for j in range(len(json_data['result'])):
            x=json_data['result'][j].get('x')
            y=json_data['result'][j].get('y')
            w=json_data['result'][j].get('w')
            h=json_data['result'][j].get('h')
            a_list.append([(x+w/2.0)*x_scale,(y+h/2.0)*y_scale,w*x_scale,h*y_scale]) # yolo 정규화
    
    f = open("C:\\Users\\user\\Desktop\\NIA_최종데이터_1207\\해충트랩\\해충\\labels\\"+i.split('.')[0]+".txt", 'w')
    for e in range(len(a_list)):
        if json_data['result'][e].get("attrb")=="총채벌레류":
            f.write("1 ")
            for r in range(len(a_list[e])):
                f.write(str(a_list[e][r]) + " ")
            f.write("\n")
        elif json_data['result'][e].get("attrb")=="목화진딧물":
            f.write("0 ")
            for r in range(len(a_list[e])):
                f.write(str(a_list[e][r]) + " ")
            f.write("\n")
        elif json_data['result'][e].get("attrb")=="왕담배나방":
            f.write("3 ")
            for r in range(len(a_list[e])):
                f.write(str(a_list[e][r]) + " ")
            f.write("\n")
        elif json_data['result'][e].get("attrb")=="작은뿌리파리":
            f.write("3 ")
            for r in range(len(a_list[e])):
                f.write(str(a_list[e][r]) + " ")
            f.write("\n")
        elif json_data['result'][e].get("attrb")=="담배거세나방":
            f.write("3 ")
            for r in range(len(a_list[e])):
                f.write(str(a_list[e][r]) + " ")
            f.write("\n")
        elif json_data['result'][e].get("attrb")=="왕담배나방":
            f.write("3 ")
            for r in range(len(a_list[e])):
                f.write(str(a_list[e][r]) + " ")
            f.write("\n")
        elif json_data['result'][e].get("attrb")=="목화바둑명나방":
            f.write("3 ")
            for r in range(len(a_list[e])):
                f.write(str(a_list[e][r]) + " ")
            f.write("\n")
        elif json_data['result'][e].get("attrb")=="진딧물류":
            f.write("0 ")
            for r in range(len(a_list[e])):
                f.write(str(a_list[e][r]) + " ")
            f.write("\n")
        elif json_data['result'][e].get("attrb")=="담배가루이":
            f.write("2 ")
            for r in range(len(a_list[e])):
                f.write(str(a_list[e][r]) + " ")
            f.write("\n")
        elif json_data['result'][e].get("attrb")=="온실가루이":
            f.write("2 ")
            for r in range(len(a_list[e])):
                f.write(str(a_list[e][r]) + " ")
            f.write("\n")
        elif json_data['result'][e].get("attrb")=="가루이류":
            f.write("2 ")
            for r in range(len(a_list[e])):
                f.write(str(a_list[e][r]) + " ")
            f.write("\n")
        elif json_data['result'][e].get("attrb")=="객체":
            f.write("4 ")
            for r in range(len(a_list[e])):
                f.write(str(a_list[e][r]) + " ")
        elif json_data['result'][e].get("attrb")=="":
#             continue
            f.write("")
        else:
            f.write("3 ")
            for r in range(len(a_list[e])):
                f.write(str(a_list[e][r]) + " ")
            f.write("\n")
    f.close()
```
