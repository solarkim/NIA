```python
import json
import os
import pprint
```

### 딸기 라벨파일 전처리

```python
with open('s3_proc/딸기/V003_3_2_1_6_4_2_2_1_0_0_20221110_9_20221115155838.json', "r",encoding='utf-8-sig') as json_file:
    json_data = json.load(json_file)
    
# print(json_data)
pprint.pprint(json_data)
```

```python
import os
root_dir = 's3/딸기/'
dir_list=os.listdir(root_dir)
dir_list
```

```python
d=0
for i in dir_list:
    with open('s3/딸기/'+i, "r",encoding="utf-8-sig",errors='ignore') as json_file:
        json_data = json.load(json_file)
        json_data['description']['time']=json_data['description']['time'].split(':')[0]+json_data['description']['time'][1]
    with open('s3_proc/딸기/'+i, 'w', encoding='utf-8-sig') as make_file:
        k5 = dict()
        k5["description"] = json_data['description']
        k5['description']['time']=json_data['description']['time'].split(':')[0]+json_data['description']['time'][1]
        k5['metadata']=json_data['metadata']
        k5['result']=json_data['result']    

        json.dump(k5, make_file,indent="\t",ensure_ascii=False)
```
